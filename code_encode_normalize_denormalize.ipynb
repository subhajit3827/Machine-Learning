```python
# prompt: read excel file in df

df = pd.read_csv('/content/drive/MyDrive/Real World Analytics Challenge/data_22_23_6.csv', low_memory=False) #Change filename.xlsx to actual name of file.
```

```python
df.head()
```

```python
# prompt: numeric encode place preferrd_foot, show table
pd.set_option('display.max_columns', None)
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['preferred_foot'] = le.fit_transform(df['preferred_foot'])
df.head()
```

```python
# prompt: decode back
pd.set_option('display.max_columns', None)
df['preferred_foot'] = le.inverse_transform(df['preferred_foot'])
df.head()
```

```python
# prompt: apply  min-max scaler to normalize overall column
pd.set_option('display.max_columns', None)
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df['overall'] = scaler.fit_transform(df['overall'].values.reshape(-1, 1))
df.head()
```

```python
# prompt: denormalize back overall column

# Assuming 'scaler' is the same MinMaxScaler object used for normalization
df['overall'] = scaler.inverse_transform(df['overall'].values.reshape(-1, 1))
df.head()
```
